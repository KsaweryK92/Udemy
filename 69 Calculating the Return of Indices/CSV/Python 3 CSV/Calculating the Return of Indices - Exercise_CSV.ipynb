## Calculating the Return of Indices

Consider three famous American market indices â€“ Dow Jones, S&P 500, and the Nasdaq for the period of 1st of January 2000 until today.

```python
import pandas
new_data = pandas.read_csv(r'/home/ksawery/PycharmProjects/Udemy/Python for Finance - Notebook Files/69 Calculating the Return of Indices/CSV/Python 3 CSV/Indices_Data_01.csv')
new_data = new_data[new_data.Date >= '2000-01-01'].set_index('Date')
print(new_data.info())
print(new_data.head())
```

Normalize the data to 100 and plot the results on a graph. 

```python
import matplotlib.pyplot as plt
((new_data / new_data.iloc[0]) * 100).plot(figsize = (15, 5))
plt.show()
```

How would you explain the common and the different parts of the behavior of the three indices?

Obtain the simple returns of the indices.

```python
((new_data / new_data.shift(1)) - 1).mean()
```

Estimate the average annual return of each index.

```python
zwroty = pandas.Series()
zwroty = ((new_data / new_data.shift(1)) - 1).mean() * 250
wagi = pandas.Series([0.25, 0.25, 0.25, 0.25])
print(zwroty)
```

```python
import numpy
numpy.dot(zwroty, wagi) * 100
```
