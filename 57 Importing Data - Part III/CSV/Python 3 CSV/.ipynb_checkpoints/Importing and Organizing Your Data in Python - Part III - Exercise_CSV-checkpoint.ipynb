Save the data obtained in the previous exercise as a csv file. Then, save it in an excel format. <br />

```python
import pandas_datareader
import pandas
```

```python
tickers = ['AAPL', 'MSFT', 'XOM', 'BP']
new_data = pandas.DataFrame()
for t in tickers:
    new_data[t] = pandas_datareader.data.DataReader(t, data_source='yahoo', start='2015-1-1')['Close']
```

```python
new_data.to_csv(r'C:\Users\Dom\Desktop\Moje\Python\Udemy\Python for Finance - Notebook Files\57 Importing Data - Part III\CSV\Python 3 CSV\new_data_moje.csv')
```

```python
new_data.to_excel(r'C:\Users\Dom\Desktop\Moje\Python\Udemy\Python for Finance - Notebook Files\57 Importing Data - Part III\CSV\Python 3 CSV\new_data_moje_EXEL.xlsx')
```
