### Simple Rate of Return

Calculate the simple returns of ‘MSFT’ for the given timeframe.

$$
\frac{P_1 - P_0}{P_0} = \frac{P_1}{P_0} - 1
$$

```python
odebrane_dane['stopa zwrotu'] = (odebrane_dane['Adj Close'] / odebrane_dane['Adj Close'].shift(1)) - 1
```
