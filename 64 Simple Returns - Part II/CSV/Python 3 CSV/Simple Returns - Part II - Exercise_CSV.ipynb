## Simple Returns - Part II

$$
\frac{P_1 - P_0}{P_0} = \frac{P_1}{P_0} - 1
$$

```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
MSFT = pd.read_csv(r'C:\Users\Dom\Desktop\Moje\Python\Udemy\Python for Finance - Notebook Files\64 Simple Returns - Part II\CSV\Python 3 CSV/MSFT_2000_2017.csv', index_col = 'Date')
MSFT['simple_return'] = (MSFT['Adj Close'] / MSFT['Adj Close'].shift(1)) - 1
print(MSFT['simple_return'])
```

Plot the simple returns on a graph.

```python
import matplotlib
```

```python
MSFT['simple_return'].plot(figsize = (20,8))
matplotlib.pyplot.show()
```

Calculate the average daily return.

```python
MSFT['simple_return'].mean()
```

Estimate the average annual return.

```python
MSFT['simple_return'].mean() * 250
```

Print the percentage version of the result as a float with 2 digits after the decimal point.

```python
print(round(((MSFT['simple_return'].mean() * 250)*100), 2), '%')
```
