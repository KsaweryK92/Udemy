## Simple Returns - Part II

$$
\frac{P_1 - P_0}{P_0} = \frac{P_1}{P_0} - 1
$$

```python
import pandas
```

```python
dane_csv = pandas.read_csv(r'/home/ksawery/PycharmProjects/Udemy/Python for Finance - Notebook Files/64 Simple Returns - Part II/CSV/Python 3 CSV/MSFT_2000_2017.csv')
```

```python
dane_csv = dane_csv.set_index('Date')
```

```python
dane_csv.tail()
```

```python
dane_csv['SIMP_RET'] = (dane_csv['Adj Close'] / dane_csv['Adj Close'].shift(1)) - 1
```

Plot the simple returns on a graph.

```python
import matplotlib.pyplot as plt
```

```python
dane_csv['SIMP_RET'].plot(figsize = (18,8))
plt.show()
```

Calculate the average daily return.

```python
daily_ret = dane_csv['SIMP_RET'].mean()
print(daily_ret)
```

Estimate the average annual return.

```python
annual_ret = daily_ret * 250
print(annual_ret)
```

Print the percentage version of the result as a float with 2 digits after the decimal point.

```python
daily_ret = round((daily_ret * 100), 2)
```

```python
annual_ret = round((annual_ret * 100), 2)
```

```python
print(f'Daily rate of return is on the level of {daily_ret}%')
```

```python
print(f'Annual rate of return is on the level of {annual_ret}%')
```
